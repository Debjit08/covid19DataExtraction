# **Collecting and Parsing a Webpage**

```
# importing the libraries
from bs4 import BeautifulSoup
import requests

url="https://www.worldometers.info/coronavirus/"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
#print(soup.prettify()) # print the parsed data of html
```

```
main_table = soup.find("table", attrs={"id": "main_table_countries_today"})
for th in main_table.find_all("th"):
  print(th.text.replace('\n', ' ').strip())
```

```
#Fetching the data
data = {}
table = main_table
t_headers = []
for th in table.find_all("th"):
  # remove any newlines and extra spaces from left and right
  t_headers.append(th.text.replace('\n', ' ').strip())
#print(t_headers)
table_data = []
for tr in table.tbody.find_all("tr"): 
  t_row={}# find all tr's from table's tbody
  for td, th in zip(tr.find_all("td"), t_headers): 
    t_row[th] = td.text.replace('\n', '').strip()
  table_data.append(t_row)
print(table_data)
```

**#Writing To CSV**

```
import csv

with open('outputdata.csv', 'w') as outfile:
    mywriter = csv.DictWriter(outfile, fieldnames=t_headers)
    mywriter.writeheader()
    for d in table_data:
        mywriter.writerow(d)
```
